```python
#import yahoo finance, import pandas. 
import yfinance as yf
import pandas as pd

#code 1 represent what you want to track.
#code 2 imports the prices for your choice for the choosen period in this case one week interval. 
#code 3 selects the price at the end of the trading period.
BTC_Ticker = yf.Ticker("BTC-USD")
BTC_Data = BTC_Ticker.history(period="max", interval="1wk")
bitcoin =BTC_Data["Close"]

ETH_Ticker = yf.Ticker("ETH-USD")
ETH_Data = ETH_Ticker.history(period="max", interval="1wk")
ethereum = ETH_Data["Close"]

AUR_Ticker = yf.Ticker("AUR-USD")
AUR_Data = AUR_Ticker.history(period="max", interval="1wk")
auroracoin = AUR_Data["Close"]

GRC_Ticker = yf.Ticker("GRC-USD")
GRC_Data = GRC_Ticker.history(period="max", interval="1wk")
gridcoin = GRC_Data["Close"]

PPC_Ticker = yf.Ticker("PPC-USD")
PPC_Data = PPC_Ticker.history(period="max", interval="1wk")
peercoin =PPC_Data["Close"]

POT_Ticker = yf.Ticker("POT-USD")
POT_Data = POT_Ticker.history(period="max", interval="1wk")
potcoin = POT_Data["Close"]

LTC_Ticker = yf.Ticker("LTC-USD")
LTC_Data = LTC_Ticker.history(period="max", interval="1wk")
litecoin=LTC_Data["Close"]

DOGE_Ticker = yf.Ticker("DOGE-USD")
DOGE_Data = DOGE_Ticker.history(period="max", interval="1wk")
dogecoin=LTC_Data["Close"]

#displaying the data
display(bitcoin)
display(ethereum)
display(auroracoin)
display(gridcoin)
display(peercoin)
display(potcoin)
display(litecoin)
display(dogecoin)
```

```python
#import matplot
import matplotlib.pyplot as plt

#we are ploting the data in the same line graph. 
plt.plot(bitcoin)
plt.plot(ethereum)
plt.plot(auroracoin)
plt.plot(gridcoin)
plt.plot(peercoin)
plt.plot(potcoin)
plt.plot(litecoin)
plt.plot(dogecoin)
plt.legend([ "Bitcoin", "Ethereum", "Auroracoin", "Gridcoin", "Peercoin", "Potcoin", "Litecoin", "Dogecoin"])
plt.show()
plt.clf()
```
